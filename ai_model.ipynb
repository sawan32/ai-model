```python
import os
import requests
import tweepy
import time
from dotenv import load_dotenv
from textblob import TextBlob
!pip install newsapi-python
from newsapi import NewsApiClient
from bs4 import BeautifulSoup

# **🔒 Load API Keys Securely**
load_dotenv()

TWITTER_BEARER_TOKEN = os.getenv("TWITTER_BEARER_TOKEN")
NEWS_API_KEY = os.getenv("NEWS_API_KEY")

# **Twitter API Authentication**
client = tweepy.Client(bearer_token=TWITTER_BEARER_TOKEN)

# **Google News Scraping (Using RSS Feed)**
def fetch_google_news(query="crypto", num_articles=5):
    url = f"https://news.google.com/rss/search?q={query}&hl=en-IN&gl=IN&ceid=IN:en"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "xml")

    news_items = []
    articles = soup.find_all("item")[:num_articles]

    for article in articles:
        headline = article.title.text if article.title else "No Title"
        link = article.link.text if article.link else "#"
        news_items.append({"headline": headline, "link": link})

    return news_items

# **Twitter Sentiment Analysis**
def analyze_twitter_sentiment(query="Bitcoin", count=10):
    try:
        tweets = client.search_recent_tweets(query=query, max_results=count)
    except tweepy.TooManyRequests:
        print("Rate limit exceeded. Waiting for 15 minutes...")
        time.sleep(15 * 60)
        tweets = client.search_recent_tweets(query=query, max_results=count)

    sentiment_scores = {"positive": 0, "neutral": 0, "negative": 0}

    if tweets.data:
        for tweet in tweets.data:
            analysis = TextBlob(tweet.text)
            if analysis.sentiment.polarity > 0:
                sentiment_scores["positive"] += 1
            elif analysis.sentiment.polarity < 0:
                sentiment_scores["negative"] += 1
            else:
                sentiment_scores["neutral"] += 1

    return sentiment_scores

# **NewsAPI Sentiment Analysis**
def fetch_newsapi_sentiment(query="cryptocurrency"):
    newsapi = NewsApiClient(api_key=NEWS_API_KEY)
    articles = newsapi.get_everything(q=query, language="en", sort_by="publishedAt", page_size=5)

    sentiment_scores = {"positive": 0, "neutral": 0, "negative": 0}

    for article in articles["articles"]:
        analysis = TextBlob(article["title"] + " " + article["description"])
        if analysis.sentiment.polarity > 0:
            sentiment_scores["positive"] += 1
        elif analysis.sentiment.polarity < 0:
            sentiment_scores["negative"] += 1
        else:
            sentiment_scores["neutral"] += 1

    return sentiment_scores

# **Combined Market Sentiment Analysis**
def combined_sentiment_analysis():
    google_news = fetch_google_news()
    twitter_sentiment = analyze_twitter_sentiment()
    newsapi_sentiment = fetch_newsapi_sentiment()

    total_positive = twitter_sentiment["positive"] + newsapi_sentiment["positive"]
    total_negative = twitter_sentiment["negative"] + newsapi_sentiment["negative"]
    total_neutral = twitter_sentiment["neutral"] + newsapi_sentiment["neutral"]

    print("\n📊 **Market Sentiment Summary**")
    print(f"✅ Positive Sentiments: {total_positive}")
    print(f"⚠️ Neutral Sentiments: {total_neutral}")
    print(f"❌ Negative Sentiments: {total_negative}")

    if total_positive > total_negative:
        print("\n📈 **Market Trend: Bullish (Consider Buying)**")
    elif total_negative > total_positive:
        print("\n📉 **Market Trend: Bearish (Consider Selling)**")
    else:
        print("\n🔍 **Market Trend: Neutral (Wait & Watch)**")

# **Run Combined Sentiment Analysis**
combined_sentiment_analysis()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ta  # Technical Analysis Library
import requests # Import the requests library

# 📌 **Load Historical Data (Replace with Live Data)**
def load_data():
    url = "https://api.binance.com/api/v3/klines?symbol=BTCUSDT&interval=1h&limit=100"
    response = requests.get(url)
    data = response.json()

    df = pd.DataFrame(data, columns=['Unix', 'Open', 'High', 'Low', 'Close', 'Volume BTC',
                                     'Close_time', 'Quote_asset_vol', 'Trades',
                                     'Taker_buy_base_vol', 'Taker_buy_quote_vol', 'Ignore'])

    df['Date'] = pd.to_datetime(df['Unix'], unit='ms')  # Convert Unix to DateTime

    # ✅ **Fix Column Names**
    df = df[['Date', 'Open', 'High', 'Low', 'Close', 'Volume BTC']]
    df.columns = ['date', 'open', 'high', 'low', 'close', 'volume']  # Rename

    return df

# Call the load_data function to create the DataFrame
df = load_data()

# 📌 **Calculate Moving Averages for Trend Confirmation**
df['EMA_50'] = df['close'].ewm(span=50, adjust=False).mean()
df['EMA_200'] = df['close'].ewm(span=200, adjust=False).mean()

# 📌 **Market Structure: Break of Structure (BOS) & Change of Character (CHoCH)**
df['Higher High'] = df['high'].shift(1) < df['high']
df['Lower Low'] = df['low'].shift(1) > df['low']

def detect_structure(df):
    df['BOS'] = (df['Higher High'] & (df['EMA_50'] > df['EMA_200'])) | (df['Lower Low'] & (df['EMA_50'] < df['EMA_200']))
    df['CHoCH'] = (df['Higher High'] & (df['EMA_50'] < df['EMA_200'])) | (df['Lower Low'] & (df['EMA_50'] > df['EMA_200']))
    return df

df = detect_structure(df)

# 📌 **Order Blocks (OB) Detection**
def find_order_blocks(df):
    df['Bullish OB'] = (df['low'] == df['low'].rolling(5).min()) & df['BOS']
    df['Bearish OB'] = (df['high'] == df['high'].rolling(5).max()) & df['BOS']
    return df

df = find_order_blocks(df)

# 📌 **Fair Value Gaps (FVG) Calculation**
def detect_fvg(df):
    df['FVG'] = (df['low'].shift(-1) > df['high']) | (df['high'].shift(-1) < df['low'])
    return df

df = detect_fvg(df)

# 📌 **Wyckoff Accumulation & Distribution Patterns**
def detect_wyckoff(df):
    df['Accumulation'] = (df['low'] == df['low'].rolling(20).min()) & (df['EMA_50'] > df['EMA_200'])
    df['Distribution'] = (df['high'] == df['high'].rolling(20).max()) & (df['EMA_50'] < df['EMA_200'])
    return df

df = detect_wyckoff(df)

# 📌 **Liquidity Zones & Stop Hunts Detection**
def detect_liquidity(df):
    df['Liquidity Zone'] = (df['high'].rolling(5).max() - df['low'].rolling(5).min()) > (df['close'].rolling(5).std() * 3)
    df['Stop Hunt'] = (df['low'] < df['low'].shift(1)) & (df['high'] > df['high'].shift(1))
    return df

df = detect_liquidity(df)

# 📌 **Order Flow Analysis using Volume**
df['Buy Volume'] = np.where(df['close'] > df['open'], df['volume'], 0)
df['Sell Volume'] = np.where(df['close'] < df['open'], df['volume'], 0)

# 📌 **Plot Data**
def plot_chart(df):
    plt.figure(figsize=(14, 7))
    plt.plot(df['date'], df['close'], label="Price", color="black")
    plt.plot(df['date'], df['EMA_50'], label="EMA 50", color="blue", linestyle="dashed")
    plt.plot(df['date'], df['EMA_200'], label="EMA 200", color="red", linestyle="dashed")
    plt.scatter(df['date'][df['BOS']], df['close'][df['BOS']], label="BOS", color="green", marker="^", alpha=0.8)
    plt.scatter(df['date'][df['CHoCH']], df['close'][df['CHoCH']], label="CHoCH", color="orange", marker="o", alpha=0.8)
    plt.scatter(df['date'][df['Bullish OB']], df['close'][df['Bullish OB']], label="Bullish OB", color="blue", marker="s", alpha=0.6)
    plt.scatter(df['date'][df['Bearish OB']], df['close'][df['Bearish OB']], label="Bearish OB", color="red", marker="s", alpha=0.6)
    plt.legend()
    plt.show()

plot_chart(df)
```
